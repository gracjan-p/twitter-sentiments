```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./twitter.csv', index_col=0)
df
```

```python
df['len'] = df['text'].apply(lambda text: len(text.split()))
plot_words = sns.displot(df['len'], bins=50, aspect=2, kde=True)
plt.xlim(0, 80)
plt.show()

from scipy.stats import hmean, trim_mean
a_mean = int(df['len'].mean())
t_mean = int(trim_mean(df['len'], 0.2))
h_mean = int(hmean(df['len']))
print(f"a_mean = {a_mean}\n"
      f"t_mean = {t_mean}\n"
      f"h_mean = {h_mean}")

df
```

```python
train = df.drop('len', axis=1)
test = df.drop('len', axis=1)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, ComplementNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), ComplementNB())
model.fit(train.text, train.label)
labels = model.predict(test.text)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score
mat = confusion_matrix(test.label, labels)
sns.heatmap(mat.T, annot=True, fmt='d', square=True, cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label')

f1 = f1_score(test.label, labels, average='weighted')
print(f1)
```
